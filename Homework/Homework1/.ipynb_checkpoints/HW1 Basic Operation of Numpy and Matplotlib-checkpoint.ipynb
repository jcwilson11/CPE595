The broadcasting rule can be also applied to the multiplication. In this question, two numpy arrays of different shapes are given, check their shape first. And then, reshape one of the array so that you can produce exactly the same result given below.

```python
A = np.array([1, 2, 3, 4])
B = np.array([[1, 2, 3, 4]])

output1 = B.shape
output2 = A.shape
A_reshaped = A.reshape(4, 1)  # Shape (4,1)
output3 = np.dot(A_reshaped, B)
print("output1: ", output1)
print("output2: ", output2)
print("output3: \n", output3)
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-whitegrid') # had to change, used print(plt.style.available)

noise_scale = 100
number_of_samples = 50
x = 30*(np.random.rand(number_of_samples, 1) - 0.5)
y = 5 * x + 15 * x**2 + 2 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='red', label="Noisy Data")
plt.title("Generated Noisy Data")
plt.xlabel("Feature 1 (X)")
plt.ylabel("Feature 2 (Y)")
plt.legend()
plt.grid(True)
plt.plot(x,y,'ro')
```

```python
# Step 1: plot the noisy data and the polynomial you found
m=4
coeffs = np.polyfit(x.flatten(), y.flatten(), m)  
print(f"Polynomial Coefficients for m={m}: {coeffs}")
x_sorted = np.sort(x, axis=0)
y_pred = np.polyval(coeffs, x_sorted)

# Plot noisy data and the found polynomial
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='red', label="Noisy Data")  
plt.plot(x_sorted, y_pred, color='blue', label=f"Polynomial Fit (m={m})")

plt.title(f"Polynomial Fit with m = 4")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Step 2: compute mean squared value
def compute_mse(x, y, m_values):
    mse_values = []
    for m in m_values:
        coeffs = np.polyfit(x.flatten(), y.flatten(), m)  
        y_pred = np.polyval(coeffs, x.flatten())  
        mse = np.mean((y.flatten() - y_pred) ** 2)  # Compute MSE
        mse_values.append(mse)
    return mse_values

m_values = list(range(1, 9))
mse_values = compute_mse(x, y, m_values)

plt.figure(figsize=(8, 6))
plt.plot(m_values, mse_values, marker='o', linestyle='-', color='blue')
plt.xlabel("Polynomial Order (m)")
plt.ylabel("Mean Squared Error (MSE)")
plt.title("MSE vs. Polynomial Order")
plt.grid(True)

plt.show()

```

```python
# Step 3: Noise scale
noise_scales = [200, 300, 400, 600, 800, 1000]
number_of_samples = 50
def plot_noise_impact(noise_scales, m):
    fig, axs = plt.subplots(2, 3, figsize=(18, 12))  # Create a 2x3 grid
    axs = axs.flatten()  # Flatten to ensure it is an iterable array
    for i, noise in enumerate(noise_scales):
            x = 30 * (np.random.rand(number_of_samples, 1) - 0.5)
            y = 5 * x + 15 * x**2 + 2 * x**3 + noise * np.random.randn(number_of_samples, 1)
    
            # Fit polynomial of the best order found in Step 2
            coeffs = np.polyfit(x.flatten(), y.flatten(), m)
            x_sorted = np.sort(x, axis=0)
            y_pred = np.polyval(coeffs, x_sorted)
    
            # Plot each subplot
            axs[i].scatter(x, y, color='red', label="Noisy Data")
            axs[i].plot(x_sorted, y_pred, color='blue', label=f"Polynomial Fit (m=8)")
            axs[i].set_title(f"Noise Scale: {noise}")
            axs[i].set_xlabel("X")
            axs[i].set_ylabel("Y")
            axs[i].legend()
            axs[i].grid(True)
    
    plt.show()

# Call function to visualize impact of noise scale
plot_noise_impact(noise_scales, m)
```

### Noise Takeaways
- when there is low noise (200, 300), the polynomial fit is accurate and smooth because the noise does not distort the pattern, when can be seen looking at the data points, which closely follow the true polynomial relationship. Overall, the lower noise levels lead to better parameter estimation and a mroe reliable fit.
- when there is medium noise (400,600), the dataset is visibly more scattered, and there are individual points that deviate further from the expected polynomial trend. Although the fit is stil reasonable, the varience in increases
- when there is high noise (800,1000), the dataset is significanly scattered, thus making it diffucult for the polynomial to capture the pattern. As a result, the polynomial coefficients are no longer reliable since they become heavily influenced by the noise. Overall, the polynomial fit becomes unstable and the estimated parameters lose accuracy.
